1. Import the necessary packages

```python
# put your answer here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

2. Load the salary.csv file

```python
# put your answer here
df = pd.read_csv("salary.csv")
df
```

3. Print the first 5 rows of the dataset

```python
# put your answer here
df.head()
```

4. Display the general information of the dataframe

```python
# put your answer here
df.info()
```

5. Visualize the data by creating a scatter plot

```python
# put your answer here
df.plot(x='YearsExperience',y='Salary',kind='scatter')
plt.show()
```

6. Split the dataset into a training set and test set

```python
# put answer here
X = df['YearsExperience'].values
y = df['Salary'].values

X = X.reshape(-1,1)
y = y.reshape(-1,1)

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

7. Create an new instance of a linear regression model

```python
# put your answer here
model = LinearRegression()
```

8. Train and test the model using the training set and test set

```python
# put your answer here
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

9. Print the equation of the line

- Display the slope
- Display the intercept
- Display the equation of the line

```python
# put your answer here
# Slope
m = model.coef_[0]
m = m[0]
# Intercept
b = model.intercept_[0]
print("b = ", b)
print("m = ", m)
print("y = ", m, "* x + ", b)
```

10. Draw the Regression Line

```python
# put your answer here
plt.scatter(X, y, color = 'blue', label='Data Points')
plt.plot(X_test, y_pred, color = 'red', linewidth=3, label = 'Regression Line')
plt.title('Relationship between Years of Experience and Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(loc=4)
plt.show()
```

11. Compute the Sum of Square Error (SSE)

```python
# put your answer here
# Reshape y_test from a two dimensional array back to a one dimensional array
y_test = y_test.reshape(-1)
# Reshape y_pred from a two dimensional array back to a one dimensional array
y_pred = y_pred.reshape(-1)

df1 = pd.DataFrame({'Actual_Sales': y_test, 'Predicted_Sales':y_pred})
df1.head()
```

```python
sum_of_squared_errors = str(np.sum(np.square(df1['Actual_Sales'] - df1['Predicted_Sales'])))
print("SSE = ", sum_of_squared_errors)
```
